## 4.0 Cálculo da variável de atraso:

#### 4.1 Filtrar pedidos com data de entrega válida


```python
orders_validas = orders[orders['order_delivered_customer_date'].notna()].copy()
```

#### 4.2 Criar coluna de atraso em dias:


```python
orders_validas['atraso_dias'] = (
    orders_validas['order_delivered_customer_date'] - 
    orders_validas['order_estimated_delivery_date']
).dt.days
```

#### 4.3 Classificar atraso como positivo (atrasado) ou não


```python
orders_validas['atrasado'] = orders_validas['atraso_dias'] > 0
```

#### 4.4 Visualizar as primeiras linhas:

```python
orders_validas[['order_id', 'order_estimated_delivery_date',
                'order_delivered_customer_date', 'atraso_dias', 'atrasado']].head()
```

## 5.0 Análise Descritiva:

#### 5.1 Distribuição das notas:

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='review_score', data=df_reviews)
plt.title('Distribuição das Notas de Avaliação')
plt.xlabel('Nota (1-5)')
plt.ylabel('Quantidade de Avaliações')
plt.show()
```

#### 5.2 Distribuição dos Atrasos:

```python
plt.figure(figsize=(8, 5))
sns.histplot(orders_validas['atraso_dias'], bins=30, kde=True, color='skyblue')
plt.title('Distribuição do Atraso em Dias')
plt.xlabel('Dias de Atraso')
plt.ylabel('Quantidade de Pedidos')
plt.show()
```

## 6.0 Relação Atraso x Nota

#### 6.1 Juntando dados de Entrega com Reviews:

```python
df_entregas_reviews = pd.merge(
    orders_validas,
    df_reviews,
    on='order_id',
    how='inner'  # apenas pedidos que têm review
)
```

#### 6.2 Visualizar as primeiras linhas


```python
df_entregas_reviews[['order_id', 'atraso_dias', 'atrasado', 'review_score']].head(3)
```

#### 6.3 Boxplot: Nota por Status de Atraso

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='atrasado', y='review_score', data=df_entregas_reviews)
plt.title('Distribuição das Notas por Status de Atraso')
plt.xlabel('Pedido Atrasado')
plt.ylabel('Nota do Review')
plt.xticks([0, 1], ['Não', 'Sim'])
plt.show()
```

#### 6.4 Média de nota por atraso

```python
nota_media = df_entregas_reviews.groupby('atrasado')['review_score'].mean().reset_index()
nota_media['atrasado'] = nota_media['atrasado'].map({False: 'Não', True: 'Sim'})

plt.figure(figsize=(6, 4))
sns.barplot(x='atrasado', y='review_score', data=nota_media)
plt.title('Nota Média por Status de Atraso')
plt.xlabel('Pedido Atrasado')
plt.ylabel('Nota Média')
plt.show()
```

#### 6.5 Correlação atraso_dias x review_score:

```python
from scipy.stats import pearsonr 

# apenas pedidos com review_score e atraso válidos
df_corr = df_entregas_reviews[
    df_entregas_reviews['atraso_dias'].notna() & 
    df_entregas_reviews['review_score'].notna()
].copy()

# Calcular correlação de Pearson
corr, p_value = pearsonr(df_corr['atraso_dias'], df_corr['review_score'])
print(f"Correlação (Pearson) entre atraso_dias e review_score: {corr:.3f}")
print(f"Valor-p: {p_value:.5f}")

```

A correlação é negativa e significativa (-0,267), indicando que atrasos maiores tendem a reduzir a nota do cliente. Entretanto, como a magnitude é baixa, outros fatores além do prazo de entrega também influenciam fortemente a satisfação.

#### 6.6  scatterplot com linha de tendência

```python

plt.figure(figsize=(8,5))
sns.regplot(
    x='atraso_dias',
    y='review_score',
    data=df_corr,
    scatter_kws={'alpha':0.3},
    line_kws={'color':'red'}
)
plt.title('Correlação entre Dias de Atraso e Nota de Avaliação')
plt.xlabel('Dias de Atraso')
plt.ylabel('Nota (1-5)')
plt.show()
```

## 7.0 Análise Regional

#### 7.1 Carregar datasets adicionais

```python
customers = pd.read_csv(os.path.join(path, 'olist_customers_dataset.csv'))
geolocation = pd.read_csv(os.path.join(path, 'olist_geolocation_dataset.csv'))
```

#### 7.2 Juntar orders com reviews e clientes

```python
df_entregas_reviews = orders.merge(df_reviews, on='order_id', how='inner')
df_entregas_reviews = df_entregas_reviews.merge(
    customers[['customer_id', 'customer_state']], 
    on='customer_id', 
    how='left'
)
```

#### 7.3 Criar coluna de atraso

```python
df_entregas_reviews['atraso_dias'] = (
    df_entregas_reviews['order_delivered_customer_date'] - 
    df_entregas_reviews['order_estimated_delivery_date']
).dt.days
df_entregas_reviews['atrasado'] = df_entregas_reviews['atraso_dias'] > 0
```

#### 7.4 Porcentagem de Atraso por Estado:

```python
atraso_estado = df_entregas_reviews.groupby('customer_state')['atrasado'].mean()
```

#### 7.5 Nota média por estado

```python
nota_estado = df_entregas_reviews.groupby('customer_state')['review_score'].mean()

print("Percentual de atrasos por estado:")
print(atraso_estado.sort_values(ascending=False))


print("\nNota média por estado:")
print(nota_estado.sort_values())
```

#### 7.6 Visualização de Atrasos por Estado:

```python
# Ordenar os estados pelo percentual de atrasos (do maior para o menor)
atraso_estado_ordenado = atraso_estado.sort_values(ascending=False)

plt.figure(figsize=(14, 7))
sns.barplot(x=atraso_estado_ordenado.index, y=atraso_estado_ordenado.values, color="steelblue")
plt.title("% de Pedidos Atrasados por Estado (Ordenado)", fontsize=16)
plt.xlabel("Estado")
plt.ylabel("Percentual de Atrasos")
plt.xticks(rotation=45)

# Adicionar os valores nas barras
for i, value in enumerate(atraso_estado_ordenado.values):
    plt.text(i, value + 0.005, f'{value:.1%}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

#### 7.7 Mapeamento estados → regiões

```python
regioes = {
    'AC': 'Norte', 'AL': 'Nordeste', 'AP': 'Norte', 'AM': 'Norte', 'BA': 'Nordeste',
    'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste', 'GO': 'Centro-Oeste',
    'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'MG': 'Sudeste',
    'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul', 'PE': 'Nordeste', 'PI': 'Nordeste',
    'RJ': 'Sudeste', 'RN': 'Nordeste', 'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte',
    'SC': 'Sul', 'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
}
```

#### 7.8 Criar coluna regiao no dataframe

```python
df_entregas_reviews['regiao'] = df_entregas_reviews['customer_state'].map(regioes)
```

#### 7.9 Calcular percentual de pedidos atrasados por região

```python
atraso_regiao = df_entregas_reviews.groupby('regiao')['atrasado'].mean().sort_values(ascending=False)
```

#### 7.10 Gráfico de barras Porcentagem de pedidos atrasados por região:

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=atraso_regiao.index, y=atraso_regiao.values, color="steelblue")
plt.title('% de Pedidos Atrasados por Região', fontsize=16)
plt.ylabel('Percentual de Atrasos')
plt.xlabel('Região')

# Adicionar valores percentuais nas barras
for i, value in enumerate(atraso_regiao.values):
    plt.text(i, value + 0.005, f'{value:.1%}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.show()
```

## 8.0 Gráficos de pizza por região(para melhor visualização)

#### 8.1 Região Nordeste

```python
df_nordeste = df_entregas_reviews[df_entregas_reviews['regiao'] == 'Nordeste']

# pedidos atrasados e não atrasados no Nordeste
contagem_nordeste = df_nordeste['atrasado'].value_counts()

# dados para o gráfico de pizza
labels = ['Não Atrasados', 'Atrasados']
sizes = [contagem_nordeste[False], contagem_nordeste[True]]
colors = ['#2ecc71', '#e74c3c']  # Verde para não atrasados, Vermelho para atrasados
explode = (0, 0.1)  # Destacar a fatia de atrasados

# gráfico de pizza
plt.figure(figsize=(10, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'fontsize': 12})

plt.title('REGIÃO NORDESTE\nDistribuição de Pedidos: Atrasados vs Não Atrasados\n' + 
          f'Total de Pedidos: {len(df_nordeste):,}'.replace(',', '.'), 
          fontsize=16, fontweight='bold', pad=20)
plt.axis('equal')  # Garante que o gráfico fique circular

# legenda com valores absolutos
legend_labels = [f'{label}: {size:,} pedidos'.replace(',', '.') for label, size in zip(labels, sizes)]
plt.legend(legend_labels, loc='best', fontsize=11)

plt.tight_layout()
plt.show()

```

| Total de pedidos           | 9.345     | 100%       |
